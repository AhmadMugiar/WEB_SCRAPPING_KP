#IMPORT LIBRARY

```python
import pandas as pd
from datetime import datetime,timedelta
import glob
from google.colab import drive
import numpy as np
from dateutil.relativedelta import relativedelta
```

#LOAD DATA

```python
drive.mount('/content/drive')
file_path = '/content/drive/MyDrive/DATA_SCRAPPING_GMAPS_POS/'
file_pattern = '*.csv'
all_files = glob.glob(file_path + file_pattern)
list_df = []
for file in all_files:
    df = pd.read_csv(file)
    list_df.append(df)
merged_df = pd.concat(list_df, ignore_index=True)
display(merged_df)
```

#PROSES CLEANING

```python
merged_df.info()
merged_df.head()
```

```python
merged_df.drop_duplicates(keep='first',inplace=True)
```

```python
for i in merged_df.columns:
    if merged_df[i].dtype=='object':
        merged_df[i]=merged_df[i].str.lower()
```

```python
merged_df['Date']=merged_df['Date'].str.replace('diedit','')
```

```python
inidict={
    'sebulan':'1 bulan',
    'setahun':'1 tahun',
    'sehari': '1 hari',
    'seminggu': '1 minggu'
}
```

```python
type(inidict),len(inidict),inidict
```

```python
merged_df['Date']=merged_df['Date'].str.replace('lalu','')
```

```python
merged_df['Date']=merged_df['Date'].str.strip()
```

```python
merged_df['Date'] = merged_df['Date'].map(inidict).fillna(merged_df['Date'])
```

```python
merged_df['Date'].value_counts()
```

```python
merged_df['waktu_dalam_angka']=merged_df['Date'].str.split(' ').str[0]
```

```python
merged_df['waktu_dalam_ket']=merged_df['Date'].str.split(' ').str[1]
```

```python
merged_df
```

```python
hariini=datetime.today().date()
```

```python
hariini
```

```python
merged_df['waktu_dalam_angka']=merged_df['waktu_dalam_angka'].astype(int)
```

```python
def hitung_tanggal(row):
    angka = row['waktu_dalam_angka']
    ket = row['waktu_dalam_ket']
    if ket == 'tahun':
        return hariini - relativedelta(years=angka)
    elif ket == 'bulan':
        return hariini - relativedelta(months=angka)
    elif ket == 'minggu':
        return hariini - relativedelta(weeks=angka)
    elif ket == 'hari':
        return hariini - relativedelta(days=angka)
    else:
        return hariini

merged_df['tanggal_review'] = merged_df.apply(hitung_tanggal, axis=1)
```

```python
merged_df.drop(columns=['Date','waktu_dalam_angka','waktu_dalam_ket'],inplace=True)
merged_df
```
