```python
from google_play_scraper import app, reviews
import pandas as pd
import numpy as np
```

```python
from logging import exception
from google_play_scraper import Sort
app_id = 'com.posindonesia.pospaymobilenew'

def get_review(app_id, lang='id', count=80000, sort=Sort.NEWEST, filter_score_with = None, filter_device_with=None, continuation_token=None):
  try :
    result, continuation_token = reviews(
      app_id,
      lang=lang,
      count=count,
      sort=sort,
      filter_score_with=filter_score_with,
      filter_device_with=filter_device_with,
      continuation_token=continuation_token
    )
    return result, continuation_token
  except Exception as e:
    print('Eror:', e)
    return None, None
reviews, continuation_token = get_review(app_id)

if reviews is not None :
  print("jumlah ulasan", len(reviews))
  if len(reviews) > 0 :
    print("ulasan pertama", reviews[0])
else :
  print("tidak ada ulasan")
```

```python
import csv

def export_to_csv(reviews, filename='sentimen_APK_PosAja.csv'):
  if reviews :
    filenames = ['reviewId', 'userName', 'rate', 'reviews_text', 'date']

    with open(filename, mode='w', newline='', encoding="utf-8") as csv_file:
      writer = csv.DictWriter(csv_file, fieldnames=filenames)
      writer.writeheader()

      for review in reviews:
        writer.writerow({
          'reviewId': review['reviewId'],
          'userName': review['userName'],
          'rate': review['score'],
          'reviews_text': review['content'],
          'date': review['at']
        })

      print (f"dara berhasil diismpan ke dalam file CSV : {filename}")
  else :
      print ("tidak ada data yang disimpan")

export_to_csv(reviews, 'sentimen_APK_Pospaymobile.csv')
```
