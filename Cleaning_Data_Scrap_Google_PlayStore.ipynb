#PROSES CLEANING

```python
merged_df = df.copy()
```

```python
merged_df.info()
merged_df.head()
```

```python
merged_df.drop_duplicates(keep='first',inplace=True)
```

```python
for i in merged_df.columns:
    if merged_df[i].dtype=='object':
        merged_df[i]=merged_df[i].str.lower()
```

```python
merged_df['date']=merged_df['date'].str.replace('diedit','')
```

```python
inidict={
    'sebulan':'1 bulan',
    'setahun':'1 tahun',
    'sehari': '1 hari',
    'seminggu': '1 minggu'
}
```

```python
type(inidict),len(inidict),inidict
```

```python
merged_df['date']=merged_df['date'].str.replace('lalu','')
```

```python
merged_df['date']=merged_df['date'].str.strip()
```

```python
merged_df['date'] = merged_df['date'].map(inidict).fillna(merged_df['date'])
```

```python
merged_df['date'].value_counts()
```

```python
merged_df['waktu_dalam_angka']=merged_df['date'].str.split(' ').str[0]
```

```python
merged_df['waktu_dalam_jam']=merged_df['date'].str.split(' ').str[1]
```

```python
merged_df
```

```python
merged_df
```

```python
merged_df.drop(columns=['date','waktu_dalam_jam'],inplace=True)
merged_df
```

```python
merged_df.rename(columns={'waktu_dalam_angka': 'tanggal reviews'}, inplace=True)
merged_df.head()
```

```python
merged_df.to_csv('Scrapping_Clean_PosPayMobile.csv',index=False)
```
